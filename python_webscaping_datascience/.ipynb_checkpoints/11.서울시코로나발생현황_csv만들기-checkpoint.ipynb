### 서울시 코로나 19 발생동향 분석
* https://www.seoul.go.kr/coronaV/coronaStatus.do
* pandas 의 read_html() 함수를 사용하여 table 엘리먼트를 추출하기
* csv 파일로 저장하기

```python
import pandas as pd

url = 'https://www.seoul.go.kr/coronaV/coronaStatus.do'

table_df = pd.read_html(url)
print(len(table_df))
table_df
```

```python
table_df[2].T
```

```python
print(table_df[3].shape)
data_df = table_df[3]
```

```python
data_df.tail()
```

```python
data_df.head()
```

```python
#마지막의 확진일 값을 파일명에 사용하기 위해서 .을 _로 변경하기
last_day = data_df.loc[0,'확진일']
print(last_day)
last_day = last_day.replace('.','_')
print(last_day)


file_name = f'data/seoul_covid_{last_day}.csv'
file_name
```

```python
#csv 파일로 저장하기
data_df.to_csv(file_name,index=False)
```

```python
#csv파일 읽기
pd.read_csv(file_name)
```

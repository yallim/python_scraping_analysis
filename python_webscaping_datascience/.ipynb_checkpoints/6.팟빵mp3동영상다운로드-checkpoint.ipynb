```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import json
import os

url = 'http://www.podbbang.com/_m_api/podcasts/16255/episodes?offset=0&sort=pubdate:desc&episode_id=0&limit=8&with=summary&cache=0'

res = requests.get(url)
print(res.status_code)
if res.ok:
    #lxml 파서를 사용해서 BeautifulSoup 객체 생성 (원래는 'html.parser')
    soup = BeautifulSoup(res.text,'lxml')
    #print(soup)
    #<p> text- </p> p태그 사이의 text 추출
    p_tag = soup.select('p')[0].text
    
    #str 타입의 텍스트를 dict 타입으로 변경하기 위해서 json 내부모듈을 사용함
    json_data = json.loads(p_tag)
    #print(type(json_data), json_data)
    
    episode_length = len(json_data['data'])
    #print(len(json_data['data']),type(json_data['data']))
    episode_list = json_data['data']
          
    for episode in episode_list:
        print('------')
        #에피소드의 제목[title]
        title = episode['title']
        #에피소드 다운로드 url
        mp3_url = episode['enclosure']['url']
        print(title,mp3_url)
        
        #url값이 있으면 다운로드 받아야 함
        if mp3_url:
            req_header = {
                'referer':'http://www.podbbang.com/ch/16255',#다운로드 요청 페이지.홈페이지 메인html
                'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.111 Safari/537.36'
            }
            
            dir_name = os.path.join('mp3','여봐라')
            print(dir_name)#mp3\여봐라
            
            #해당 디렉토리가 없으면
            if not os.path.isdir(dir_name):
                #디렉토리를 생성
                os.makedirs(dir_name)
            
            #해당 url에서 파일명만 추출
            file_name = os.path.basename(mp3_url)
            #생성된 디렉토리명과 파일명을 합친다.
            file_name = dir_name+'/'+file_name
            
            #print(file_name)
            res = requests.get(mp3_url,headers=req_header)
            if res.ok:
                #response 객체에서 binary데이터 추출
                mp3_bin = res.content
                #binary 데이터를 local 파일로 저장
                with open(file_name,'wb') as file:
                    file.write(mp3_bin)
            
```

### 팟빵 다운로드 1단계
* 팟빵 id와 제목을 함수의 인자로 받는다.
* 첫번째의 8개의 에피소드만 다운로드 받고 있다.
* 저장되는 mp3 파일명을 에피소드의 타이틀로 설정함

```python
#문자열에 있는 특수문자를 제거하는 함수
def clean_text(text):
    import re
    text = text.replace("\n", "")

    cleaned_text = re.sub('[\{\}\[\]\/?.,;:|\)*~`!^\-_+<>@\#$%&\\\=\(\'\"]','',text)
    cleaned_text = cleaned_text.replace(' ','')

    return cleaned_text
```

```python
clean_text("김영철의 파워FM - 진짜 미국식 영어 810회 - 타일러의 진짜 미국식 영어 = Don't be so stubborn. = 억지 좀 부리지 마세요!! ")
```

```python
def get_podbbang(input1,input2):
    import requests
    from bs4 import BeautifulSoup
    from urllib.parse import urljoin
    import json
    import os
    
    url = f'http://www.podbbang.com/_m_api/podcasts/{input1}/episodes'
    req_param= {
        "offset":0,
        "sort":"pubdate:desc",
        "episode_id":0,
        "limit":8,
        "with":"summary",
        "cache":"0"
    }

    res = requests.get(url,params=req_param)
    print(res.status_code)
    if res.ok:
        #lxml 파서를 사용해서 BeautifulSoup 객체 생성 (원래는 'html.parser')
        soup = BeautifulSoup(res.text,'lxml')
        #print(soup)
        #<p> text- </p> p태그 사이의 text 추출
        p_tag = soup.select('p')[0].text

        #str 타입의 텍스트를 dict 타입으로 변경하기 위해서 json 내부모듈을 사용함
        json_data = json.loads(p_tag)
        #print(type(json_data), json_data)
        
        
        episode_length = len(json_data['data'])
        #print(len(json_data['data']),type(json_data['data']))
        episode_list = json_data['data']

        for episode in episode_list:
            print('------')
            #에피소드의 제목[title]
            episode_title = episode['title']
            #에피소드 다운로드 url
            mp3_url = episode['enclosure']['url']
            print(episode_title,mp3_url)

            #url값이 있으면 다운로드 받아야 함
            if mp3_url:
                req_header = {
                    'referer':f'http://www.podbbang.com/ch/{input1}',#다운로드 요청 페이지.홈페이지 메인html
                    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.111 Safari/537.36'
                }

                dir_name = os.path.join('mp3',input2)
                print(dir_name)#mp3\여봐라

                #해당 디렉토리가 없으면
                if not os.path.isdir(dir_name):
                    #디렉토리를 생성
                    os.makedirs(dir_name)

                #해당 url에서 파일명만 추출
                #file_name = os.path.basename(mp3_url)
                episode_title = clean_text(episode_title)
                file_name = f'{episode_title}.mp3'
                
                #생성된 디렉토리명과 파일명을 합친다.
                file_name = dir_name+'/'+file_name

                #print(file_name)
                res = requests.get(mp3_url,headers=req_header)
                if res.ok:
                    #response 객체에서 binary데이터 추출
                    mp3_bin = res.content
                    #binary 데이터를 local 파일로 저장
                    with open(file_name,'wb') as file:
                        file.write(mp3_bin)
```

### 팟빵 다운로드 2단계
* 팟빵 id와 제목을 함수의 인자로 받는다.
* 첫번째의 8개의 에피소드만 다운로드 받고 있다.
* 저장되는 mp3 파일명을 에피소드의 타이틀로 설정함
 - 에피소드를 요청할때 offset 값의 최대값이 항상 변하므로
 - itertools의 count()함수를 사용해 무한루프를 수행하도록 한다.
 - 무한루프를 빠져나올 수 있는 조건식을 반드시 주어야 한다.
   - json_data['data'] 길이가 0이면 무한루프를 탈출한다.
 - 사람이 직접 다운로드 받는 것 처럼 보이게 하기 위해서 sleep time을 설정
   - time 이라는 내부모듈의 sleep 함수를 사용한다.

```python
def get_all_podbbang(input1,input2):
    import requests
    from bs4 import BeautifulSoup
    from urllib.parse import urljoin
    import json
    import os
    from itertools import count
    from time import sleep
    
    url = f'http://www.podbbang.com/_m_api/podcasts/{input1}/episodes'

    for offset in count(0):
        
        req_param= {
            "offset":offset,
            "sort":"pubdate:desc",
            "episode_id":0,
            "limit":8,
            "with":"summary",
            "cache":"0"
        }

        res = requests.get(url,params=req_param)
        print(res.status_code)
        if res.ok:
            #lxml 파서를 사용해서 BeautifulSoup 객체 생성 (원래는 'html.parser')
            soup = BeautifulSoup(res.text,'lxml')
            #print(soup)
            #<p> text- </p> p태그 사이의 text 추출
            p_tag = soup.select('p')[0].text

            #str 타입의 텍스트를 dict 타입으로 변경하기 위해서 json 내부모듈을 사용함
            json_data = json.loads(p_tag)
            #print(type(json_data), json_data)


            episode_length = len(json_data['data'])
            print('offset:',offset,'data의 길이=>',episode_length)
            #무한루프 탈출구문
            if episode_length ==0:
                return None
            
            #print(len(json_data['data']),type(json_data['data']))
            episode_list = json_data['data']

            for episode in episode_list:
                print('------')
                #에피소드의 제목[title]
                episode_title = episode['title']
                #에피소드 다운로드 url
                mp3_url = episode['enclosure']['url']
                print(episode_title,mp3_url)

                #url값이 있으면 다운로드 받아야 함
                if mp3_url:
                    req_header = {
                        'referer':f'http://www.podbbang.com/ch/{input1}',#다운로드 요청 페이지.홈페이지 메인html
                        'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.111 Safari/537.36'
                    }

                    dir_name = os.path.join('mp3',input2)
                    print(dir_name)#mp3\여봐라

                    #해당 디렉토리가 없으면
                    if not os.path.isdir(dir_name):
                        #디렉토리를 생성
                        os.makedirs(dir_name)

                    #해당 url에서 파일명만 추출
                    #file_name = os.path.basename(mp3_url)
                    episode_title = clean_text(episode_title)
                    file_name = f'{episode_title}.mp3'

                    #생성된 디렉토리명과 파일명을 합친다.
                    file_name = dir_name+'/'+file_name

                    #print(file_name)
                    res = requests.get(mp3_url,headers=req_header)
                    if res.ok:
                        #response 객체에서 binary데이터 추출
                        mp3_bin = res.content
                        #binary 데이터를 local 파일로 저장
                        with open(file_name,'wb') as file:
                            file.write(mp3_bin)
                            
                        #0.5초간 프로세스를 중지 => 기계가 아니라 사람처럼 보이게 하기위해서임.
                        sleep(0.5)
                            
        
```

```python
get_all_podbbang('16255','여바라')
```
